```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('fifa-world-cup-data\WorldCups.csv')
df.head(11)
```

```python
sns.relplot(x="Year", y="GoalsScored", data=df)
plt.title('FIFA World Cup')
plt.show()
```

```python
xlabels = df.Year
ax = sns.lineplot(x="Year", y="GoalsScored", data=df)
ax.set_title('FIFA World Cup')
ax.set_xticklabels(xlabels)

plt.show()
```

```python
xlabels = df.Country
ax = sns.scatterplot(x="Country", y="GoalsScored", data=df, )
ax.set_title('FIFA World Cup')
ax.set_xticklabels(xlabels, rotation=45)

plt.show()
```

```python
# Regression plots
ax = sns.lmplot(x="Year", y="GoalsScored", data=df)
plt.show()
```

```python
ax = sns.PairGrid(df)
ax.map_diag(sns.kdeplot)
ax.map_offdiag(sns.kdeplot, n_levels=6);
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.Year, df.GoalsScored, cmap=cmap, n_levels=60, shade=True);
```

```python
sns.pairplot(df)
plt.show()
```
