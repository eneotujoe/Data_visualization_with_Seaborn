```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
sns.set()
```

```python
iris_data = sns.load_dataset("iris")
```

```python
iris_data.head()
```

```python
iris_data.columns
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris_data, kind='hex',color='yellow')
plt.savefig('jointplot')
plt.show()
```

```python
# Relational plot
ax = sns.relplot(x="sepal_length", y="sepal_width", hue="species", data=iris_data)

plt.show()
```

```python
# scatterplot
ax = sns.scatterplot(x="petal_length", y="petal_width", data=iris_data, hue="species", palette="Set1")
plt.show()
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", data=iris_data)
plt.show()
```

```python
# line plot 
ax = sns.lineplot(x="sepal_length", y="sepal_width", data=iris_data, hue="species", palette="Set3")
plt.show()
```

```python
sns.pairplot(iris_data, hue="species", palette="Set3")
plt.show()
```

```python
# Heatmap
corr = iris_data.corr()
sns.heatmap(corr,annot=True,cmap='YlGnBu') 
plt.show()
```

```python
xlabels = iris_data.petal_length
ax = sns.relplot(x="petal_length", y="petal_width", hue="species", data=iris_data)
ax.set_xticklabels(xlabels)

plt.show()
```
