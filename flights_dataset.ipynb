```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style("dark")
```

```python
flights_data = sns.load_dataset("flights")
```

```python
flights_data.head()
```

```python
flights_data.columns
```

```python
sns.jointplot(x="year", y="passengers", data=flights_data, kind='hex',color='yellow')
plt.savefig('jointplot')
plt.show()
```

```python
# Relational plot
ax = sns.relplot(x="year", y="passengers", hue="month", data=flights_data)

plt.show()
```

```python
# scatterplot
ax = sns.scatterplot(x="year", y="passengers", data=flights_data, hue="month", palette="Set1")
plt.show()
```

```python
sns.lmplot(x="year", y="passengers", data=flights_data)
plt.show()
```

```python
# line plot 
ax = sns.lineplot(x="year", y="passengers",data=flights_data, hue="month", palette="Set3")
plt.show()
```

```python
sns.pairplot(flights_data, hue="month", palette="Set3")
plt.show()
```

```python
# Heatmap
corr = flights_data.corr()
sns.heatmap(corr,annot=True,cmap='YlGnBu') 
plt.show()
```

```python
xlabels = flights_data.month
ax = sns.relplot(x="year", y="passengers", hue="year", data=flights_data)
ax.set_xticklabels(xlabels, rotation = 45)

plt.show()
```
