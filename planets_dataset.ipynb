```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style("dark")
```

```python
planets_data = sns.load_dataset("planets")
```

```python
planets_data.head(17)
```

```python
planets_data.columns
```

```python
sns.jointplot(x="year", y="mass", data=planets_data, kind='hex',color='yellow')
plt.savefig('jointplot')
plt.show()
```

```python
# Relational plot
ax = sns.relplot(x="year", y="mass", hue="number", data=planets_data)

plt.show()
```

```python
# scatterplot
ax = sns.scatterplot(x="year", y="mass", data=planets_data, hue="number", palette="Set3")
plt.show()
```

```python
sns.lmplot(x="year", y="mass", data=planets_data)
plt.show()
```

```python
# line plot 
ax = sns.lineplot(x="year", y="mass", data=planets_data, hue="number", palette="Set3")
plt.show()
```

```python
# Heatmap
corr = planets_data.corr()
sns.heatmap(corr,annot=True,cmap='YlGnBu') 
plt.show()
```

```python
xlabels = planets_data.year
ax = sns.relplot(x="year", y="mass", hue="method", data=planets_data)
ax.set_xticklabels(xlabels)

plt.show()
```
