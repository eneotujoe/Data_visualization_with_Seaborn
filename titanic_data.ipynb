```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style("dark")
```

```python
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.head(10)
```

```python
titanic_data.columns
```

```python
sns.jointplot(x="age", y="fare", data=titanic_data, kind='hex',color='yellow')
plt.savefig('jointplot')
plt.show()
```

```python
# Relational plot
ax = sns.relplot(x="age", y="fare", hue="sex", data=titanic_data)

plt.show()
```

```python
# scatterplot
ax = sns.scatterplot(x="age", y="fare", data=titanic_data, hue="sex", palette="Set2")
plt.show()
```

```python
sns.lmplot(x="age", y="fare", data=titanic_data)
plt.show()
```

```python
# line plot 
ax = sns.lineplot(x="age", y="fare", data=titanic_data, hue="sex", palette="Set2")
plt.show()
```

```python
# Heatmap
corr = titanic_data.corr()
sns.heatmap(corr,annot=True,cmap='YlGnBu') 
plt.show()
```

```python
xlabels = titanic_data
ax = sns.relplot(x="age", y="fare", hue="sex", data=titanic_data)
ax.set_xticklabels(xlabels)

plt.show()
```
